```python
# Converting the date from DD/MM/YYYY to YYYY-MM-DD
date_info = pd.to_datetime(air_quality_['Date'], format='%d/%m/%Y', dayfirst=True)

# Now, if you want to format it as YYYY-MM-DD (which is the default in pandas)
air_quality_['Date'] = date_info.dt.strftime('%Y-%m-%d')

print(air_quality_['Date'].head())
```

```python
time_info= air_quality_['Time']
```

```python
print(time_info)
```

```python
time_info = time_info.apply(lambda x: x.replace('.', ':'))
```

```python
print(time_info)
```

```python
print(type(date_info))
print(type(time_info))
```

```python
#combining / concating both 2 series to a pandas dataframe
date_time = pd.concat([date_info,time_info],axis=1)
date_time.head()
```

```python
#combining date and time
date_time['ds']= date_time['Date'].astype(str)+' '+ date_time['Time'].astype(str)
```

date and time should be combine as 'ds' with type as 'String'

```python
date_time.head()
```

```python
date_time.info()
```

Now, we have to covert ds from object to date time format

```python
data= pd.DataFrame()
```

```python
data['ds']= pd.to_datetime(date_time['ds'])
```

```python
data.head()
```

```python
data['y']= air_quality_['RH']
```

```python
data.head()
```

Now, we need to install library called 'prophet'

```python
from prophet import Prophet
```

```python
#Traning the prophet model

model= Prophet()
model.fit(data)
```

```python
# Generate future dataframe 
future = model.make_future_dataframe(periods=365,freq='H')

# Make predictions
forecast = model.predict(future)

# Display the last few predictions to verify the date range
print(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail())
```

```python
fig1 = model.plot(forecast)
```

```python
fig2 = model.plot_components(forecast)
```
